### Plot the distributions of two classes, 1. GDP < 58000 and  2. GDP >= 58000, of COVID-19 State Data using different colors and different  markers where x-axis is the Pollution and y-axis the Mortality-rate.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties # 步驟一
```

```python
#設定兩種條件的資料:

#GDP < 58000
df_GDP1 = df[df["GDP"] < 58000]
df_GDP1_only = df_GDP1["GDP"]
#print(df_GDP1_only)

#GDP >= 58000
df_GDP2 = df[df["GDP"] >= 58000]
df_GDP2_only = df_GDP1["GDP"]
#print(df_GDP2_only)

#畫圖:

#折線圖
print("Line graph: ")
plt.plot(df_GDP1["Pollution"], df_GDP1["Deaths"]/df_GDP1["Infected"])
plt.plot(df_GDP2["Pollution"], df_GDP2["Deaths"]/df_GDP2["Infected"])

plt.title('COVID-19 Data')  #圖表標題
plt.xlabel('Pollution')  #x軸說明文字
plt.ylabel('Mortality-rate')  #y軸說明文字

plt.show()
  
#散佈圖
print("Scatter Plot:  ")
plt.scatter(df_GDP1["Pollution"], df_GDP1["Deaths"]/df_GDP1["Infected"]) #XY軸參數為汙染與死亡率
plt.scatter(df_GDP2["Pollution"], df_GDP2["Deaths"]/df_GDP2["Infected"])

plt.title('COVID-19 Data')  #圖表標題
plt.xlabel('Pollution')  #x軸說明文字
plt.ylabel('Mortality-rate')  #y軸說明文字

plt.show()
```

# Step 7
### Show the proportion of following three classes of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1、 Mortality-rate < 0.02 <p>
 2、 Mortality-rate between 0.02 and 0.03 <p>
 3、Mortality-rate > 0.03 <p>

```python
#設定個別的資料數量:

#print("M1")
df_MR1 = df[df["Deaths"]/df["Infected"] < 0.02] #存取滿足條件的資料
num_df_MR1 = len(df_MR1.index) #紀錄資料數量


#print("M2")
df_MR2 = df[df["Deaths"]/df["Infected"] >= 0.02]
df_MR2 = df_MR2[df_MR2["Deaths"]/df_MR2["Infected"] < 0.03]
num_df_MR2 = len(df_MR2.index)

#print("M3")
df_MR3 = df[df["Deaths"]/df["Infected"] > 0.03]
num_df_MR3 = len(df_MR3.index)

#標籤分別對應至上面的M1 M2 M3
labels = 'Mortality-rate < 0.02', 'Mortality-rate between 0.02 and 0.03', 'Mortality-rate > 0.03' 

#Size為每塊的大小(使用資料數量來畫圖)
size = [num_df_MR1, num_df_MR2, num_df_MR3]

#autopct='%1.1f%%'是用來顯示百分比
plt.pie(size , labels = labels,autopct='%1.1f%%')

#為了要讓圓餅圖比例相等加上
plt.axis('equal')

plt.show()


```
